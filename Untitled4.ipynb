```python
#list 선언
def cncn():
    index = []
    title_list = []
    contents_list = []
    hearthy_tag_list  = []
    no = 1


    for j in li_list:
        try : 
            title = j.find('div','tit').get_text().strip()
            
        except:
            continue
        else:
            # 각 변수명으로 추출
            contents = j.find('div','service').get_text().strip()
            hearthy_tag= j.find('p','tag_type').get_text().strip()
           

            # 원하는 모양으로 출력
            print('1. 제목: ', title)
            print('2. 지역: ', contents)
            print('3. 해시태그: ', hearthy_tag)
            

            # 리스트에 저장
            title_list.append(title)
            contents_list.append(contents)
            hearthy_tag_list.append(hearthy_tag)

        print("\n")
        print("="*100)
```
