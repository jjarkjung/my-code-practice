```python

index = []
title_list = []
star_score_list = []   
review_list = []
```

```python
#list 선언
def akak():
    full_html = driver.page_source # 현재 페이지의 HTML 전체 코드를 지정
    soup = BeautifulSoup(full_html, 'html.parser') #위의 전체 HTML 코드를 가지고 온다
    content_list = soup.find('body')
    div_list = content_list.find_all('div',class_='article')
    li_list = content_list.find_all('ul')
    
    
        

  # 리스트에 저장       

    
    for q in div_list:
        title = q.find('strong',class_='h_lst_tx').get_text().strip()
        star_score = q.find('div',class_='top_behavior').get_text().strip()
        review = q.find('div',class_="user_tx_area").get_text().strip()            
        review_list.append(review)
        title_list.append(title)
        star_score_list.append(star_score)    
        print('1. 제목: ', title)
        print('2. 별점: ', star_score)
        print('3. 리뷰: ', review)
        
        
        
        print("\n")
        print("="*100)
```

```python
def create_folder(keyword):
    #현재 시간 저장하기
    now = time.localtime() #현재시간 now로 저장하기
    s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

    f_dir = 'c:\\temp\\' #입력받은 경로

    print(f_dir+keyword+'-'+s)

    import os

    os.makedirs(f_dir+keyword+'-'+s) 
    os.chdir(f_dir+keyword+'-'+s)
    # 크롤링 결과 저장할 폴더 이름
    f_dir_final = f_dir+keyword+'-'+s
    return f_dir_final 
```

```python
import pandas as pd
def ddaa():

    df = pd.DataFrame()
    df['제목'] = title_list
    df['평점'] = star_score_list
    df['리뷰'] = review_list


    f_name = "c:\\temp\\movie.csv"
    f_name_xls = "c:\\temp\\movie.xls"
    df.to_csv(f_name, encoding = 'utf-8-sig')
    df.to_excel(f_name_xls, encoding = 'utf-8-sig')
```

```python
cnt=30000
page_cnt = math.ceil(cnt / 10)

no = 1

for x in range(1,page_cnt+1) : 
    print("%s 페이지 내용 수집 시작합니다 =======================" %x)
    print("\n")    
    time.sleep(2)          
    
    
    for title_no in range(1,11):
        time.sleep(1)
        
        if no-1 >= cnt:
                break
        try:
    
            driver.find_element_by_css_selector('#reviewTab > div > div > ul > li:nth-child(%s) > a > strong' %title_no).click( )
            time.sleep(2)
            print("article no:", no)
            akak()
            ddaa()
            
            print("page_for loop iteration : " , title_no)

        except:
            continue
        print("driver.back")
        driver.back()
        print("driver.back_2")
        
    x += 1
    driver.find_element_by_link_text("""%s""" %x).click()     
        
print("크롤링 완료!")
```
